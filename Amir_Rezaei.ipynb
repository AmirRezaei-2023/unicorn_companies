```python
new_list=[]
for i in mylist:
    if i== None:
        pass
    else:
        new_list.append(i)
print(new_list)
```

```python
n_list=[]
for i in new_list:
    ms=str(i)
    j=ms.replace(' ','')
    n_list.append(j)
print(n_list)
```

```python
my_dict = {}
for i in n_list:
    #print(i)
    my_dict[i] = my_dict.get(i, 0) + 1
print(my_dict)
```

**19-In your opinion, why is there still no unicorn in Iran?**

**Ans** : In my opinion, there could be a few reasons why there may not yet be a unicorn company in Iran:

1. Economic and political factors: Iran's economy has faced various challenges over the years, including international sanctions, which can hinder the growth of startups and access to funding. Political and regulatory limitations may also impact the ability of companies to scale and attract investment.

2. Limited access to venture capital: The availability of venture capital plays a crucial role in the growth of unicorn companies. In Iran, the venture capital ecosystem may still be developing, with limited funding options and support for high-growth startups. This shortage of venture capital can make it difficult for companies to secure the large investments needed to reach unicorn status.

3. Market size and competition: While Iran has a sizable population, the domestic market may not be as large or as competitive compared to some other countries. Unicorn companies often require significant market demand and scalability, which may be challenging in a relatively smaller market.

4. Startup ecosystem maturity: Building a successful startup ecosystem takes time, and Iran's startup ecosystem may still be in the early stages of development. A robust ecosystem requires supportive infrastructure, mentorship, networking opportunities, and a culture that encourages entrepreneurship and risk-taking. It may take time for these elements to fully develop and support the emergence of unicorn companies.

**It's important to note that these are potential reasons and not definitive explanations. The dynamics surrounding the growth of unicorns can vary significantly depending on the country and its unique circumstances. Over time, as factors change and evolve, it is possible that we may see the emergence of unicorn companies in Iran.**
